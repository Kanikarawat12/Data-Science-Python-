### Which Year has Maximum Export and Import

```python
df1=df.groupby('Year').sum()
df1
```

```python
df1.info() 
```

```python
%matplotlib inline
Year = range(2015,2022)
plt.bar(Year,df1['Value'])
plt.xlabel('Year')
plt.ylabel('Import/Export')
plt.title('Year with Maximum Import/Export')
plt.show()
```

### 2021 has Maximum Import and Export

### Which month of 2021 has maximum export/import?

### Add a Month Column


# lets use the .apply() method

```python
df['Month'] = df['Date'].apply(lambda x: x.split('/')[1])
df
```

### Now Make a Subset of 2021 Rows

```python
df2021 = df[df['Year']>2020]
df2021
```

```python
df_months = df2021.drop(['Year'] , axis = 1)
df_months
```

```python
df_2021 = df_months.groupby('Month').sum()
df_2021
```

```python
%matplotlib notebook
Months = range(1,13)
plt.bar(Months, df_2021['Value'])
plt.xlabel('Month')
plt.ylabel('Import/Export')
plt.title('Import/Export in 2021')
plt.show()
```

### November has maximum import/export 

### December has minimum import/export
