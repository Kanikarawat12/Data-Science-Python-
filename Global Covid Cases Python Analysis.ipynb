### We will make separate subsets of cases in January month of 2020, 2021 and 2022.

```python
df_jan2020=df.loc[0:28]
df_jan2020
```

```python
df_jan2021 = df.loc[364:394]
df_jan2021
```

```python
df_jan2022=df[729:757]
df_jan2022
```

```python
df_afgan = pd.concat([df_jan2020,df_jan2021,df_jan2022])
df_afgan
```

```python
df_jan_yearly = df_afgan.groupby('Year').sum()
df_jan_yearly
```

### LET US DRAW GRAPHS OF OBSERVATIONS

```python
Year = [2020,2021,2022]
plt.plot(Year , df_jan_yearly['New_cases'])
plt.title('AFGANISTAN January Cases')
plt.xlabel('Year')
plt.ylabel('New_cases')
plt.show()
```

```python
df_who_region= df.groupby('WHO_region').sum()
df_who_region
```

```python
WHO_Region = pd.unique(df['WHO_region'])
plt.plot(WHO_Region, df_who_region['New_cases'])
plt.title('Cases In WHO Region')
plt.xlabel('WHO REGION')
plt.ylabel('CASES')
plt.show()
```

### Let's see Country Wise Cases

```python
df_countrywise = df.groupby('Country').sum()
df_countrywise
```

```python
%matplotlib inline
Country = pd.unique(df['Country'])
plt.plot(Country, df_countrywise['New_cases'])
plt.xlabel('COUNTRY')
plt.ylabel('Cases')
plt.title('Country-wise Cases')
plt.show()
```
